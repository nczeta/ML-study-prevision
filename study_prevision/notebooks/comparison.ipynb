# Model Comparison

In this notebook we are going to compare the linear regression model and the decision tree model built in the src folder. We are going to analize the errors of each model and make some conclusions.

First of all let's analize the errors calculated in the script `regression.py`.

The Mean Error (ME) measures the average difference between actual and predicted values. If the error is equal to 0 the model is perfect. If `me > 0`, the model is overestimating the prediction so `actual_value < prediction`. If `me < 0`, the model is underestimating , `actual_values > prediction`. In this case, the linear regression is underestimating the predictions.

```python
me_regression = -0.029779712847401243
```

The Mean Absolute Error (MAE) measures the average of the absolute differences between actual and predicted values, ignoring the sign. It tells you how much your model is off on average, making it more reliable than ME since errors don't cancel each other out.The higher it is, the less precise the model is.

```python
mae_regression = 0.41201926802736943
```

The Mean Squared Error (MSE) measures the average of the squared differences between actual and predicted values. Because errors are squared, larger mistakes are penalized much more heavily than smaller ones, making it more sensitive to outliers than MAE. A low MSE indicates the model predicts closely to the actual values, while a high MSE signals large errors. Since values are squared, the MSE is not in the same unit as your data, making it harder to interpret directly 

```python
mse_regression = 0.26673251852923635
```
